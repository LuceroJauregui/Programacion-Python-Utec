**2. Manipulación de listas (1 puntos)**

**3. Operaciones con listas (2 puntos)**

Crea una lista llamada resultante que contenga la raíz cuadrada de cada
elemento de la lista numeros_pares usando list comprehension.

```python
import math
# Crear nuevamente la lista de números pares del 2 al 100
numeros_pares = list(range(2, 101, 2))

# Crear la lista resultante con raíces cuadradas redondeadas a 2 decimales
resultante = [round(math.sqrt(num), 2) for num in numeros_pares]

# Mostrar la lista resultante
print("resultante =", resultante)
```

**5. Iteración en diccionarios (4 puntos)**

Dado el siguiente diccionario notas = {“Programacion”:15.9,
“Matematica”: 15.1, “Fisica”:15.6}, escribe un bucle for que imprima
cada clave y su valor en el diccionario con la siguiente estructura “Mis
calificaciones en el curso de Programacion es 15.9”


```python
# Diccionario de notas
notas = {"Programacion": 15.9, "Matematica": 15.1, "Fisica": 15.6}

# Bucle para iterar e imprimir
for curso, nota in notas.items():
    print(f"Mis calificaciones en el curso de {curso} es {nota}")

# Calcular el promedio
promedio = sum(notas.values()) / len(notas)

# Imprimir el promedio
print(f"Mi promedio general es {round(promedio, 2)}")

# Evaluar si está aprobado o desaprobado
if promedio >= 14: print("Estado: Aprobado ✅")
else: print("Estado: Desaprobado ❌")
```

**6. Matrices y funciones de NumPy (4 puntos)**

Crea una matriz 1D que contenga los primeros 16 elementos de la lista
numeros_pares (creada en la pregunta 1), modifique las dimensiones y
convertiela en una matriz 2D con shape(4,4), calcula la suma total de los
elementos de la matriz matriz 2D.

```python
import numpy as np
# Crear nuevamente la lista de la pregunta 1
numeros_pares = list(range(2, 101, 2))

# Tomar los primeros 16 elementos
primeros_16 = numeros_pares[:16]

# Crear una matriz 1D con NumPy
matriz_1d = np.array(primeros_16)

# Cambiar la forma a (4,4) para obtener matriz 2D
matriz_2d = matriz_1d.reshape(4, 4)

# Calcular la suma total de los elementos
suma_total = matriz_2d.sum()

# Mostrar resultados
print("Matriz 2D:\n", matriz_2d)
print("Suma total de elementos:", suma_total)
```

**7. Pandas (6 puntos)**

Leer el archivo datosDemograficos.xlsx y realice un análisis descriptivo en
función de las variables más relevantes.

```python
# Conectando con Drive y cargando el Dataset
from google.colab import drive
drive.mount('/gdrive')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar Dataset
data_ruta= ('/content/datosDemograficos (1).xlsx')
dataset=pd.read_excel(data_ruta)

dataset.head()
```

**1. Estadísticas Descriptivas Básicas**

```python
# Resumen estadístico de variables numéricas
dataset.describe()
```

*Verificación nombres de columnas*

```python
dataset.columns = dataset.columns.str.strip()
print(dataset.columns)
```

**2. Matriz de Correlación**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

✅ ***Conclusiones Clave***

La esperanza de vida está fuertemente afectada por la mortalidad infantil, lo cual es coherente y útil para políticas de salud pública.

El crecimiento natural depende casi enteramente de la fecundidad, lo cual indica que aún hay regiones con alta natalidad.


**3. Boxplot para detectar outliers**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=dataset[["Fecundidad", "Mortalidad", "TasaMortalidadxMil","EsperanzaVida", "TasaMortalidadInfantilxMilNacidos","CrecimientoNatural"]])
plt.xticks(rotation=45)
plt.title("Distribución de Variables Demográficas")
plt.show()
```

***✅Conclusiones clave***

Lima es un outlier en casi todas las variables absolutas (fecundidad, mortalidad, crecimiento natural). Esto afecta la escala del gráfico y puede distorsionar visualmente las diferencias entre regiones más pequeñas.

Las tasas (por mil o porcentajes) son mejores para comparar entre regiones, ya que no dependen del tamaño poblacional.

La mortalidad infantil aún es preocupante en algunas regiones, dado que tiene varios valores elevados fuera del rango esperado.

La esperanza de vida es bastante uniforme, lo cual refleja políticas nacionales homogéneas en salud, aunque hay pequeños márgenes de mejora en algunas zonas.

**4. Relación entre Fecundidad y Esperanza de Vida**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x="Fecundidad", y="EsperanzaVida", hue="Indicadores")
plt.title("Relación: Fecundidad vs Esperanza de Vida")
plt.xlabel("Fecundidad")
plt.ylabel("Esperanza de Vida")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**5. Top 5 regiones con mayor crecimiento natural**

```python
# Ordenar por crecimiento natural descendente
top_crecimiento = dataset.sort_values(by="CrecimientoNatural", ascending=False)

# Mostrar en consola el top 5
print("Top 5 regiones con mayor crecimiento natural:")
print(top_crecimiento[["Indicadores", "CrecimientoNatural"]].head())

# Crear gráfico de barras del top 5
plt.figure(figsize=(8, 5))
sns.barplot(data=top_crecimiento.head(5),x="CrecimientoNatural",y="Indicadores",palette="crest")

plt.title("Top 5 Regiones con Mayor Crecimiento Natural")
plt.xlabel("Crecimiento Natural")
plt.ylabel("Región")
plt.tight_layout()
plt.show()
```

**6. Clasificación por cuartiles y gráfico**

```python
# Crear una columna que clasifique la esperanza de vida en cuartiles
dataset['Cuartil_EV'] = pd.qcut(dataset['EsperanzaVida'], q=4, labels=["Baja", "Media-Baja", "Media-Alta", "Alta"])

# Contar cuántas regiones hay por cada cuartil
cuartiles_count = dataset['Cuartil_EV'].value_counts().sort_index()

# Gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(cuartiles_count, labels=cuartiles_count.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Distribución de Regiones según Cuartiles de Esperanza de Vida")
plt.axis('equal')  # Mantener forma de círculo
plt.show()
```

**7. Dispersión bivariada (Fecundidad vs Tasa Mortalidad Infantil)**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='Fecundidad', y='TasaMortalidadInfantilxMilNacidos', hue='Indicadores')
plt.title("Fecundidad vs Tasa de Mortalidad Infantil por Región")
plt.xlabel("Fecundidad")
plt.ylabel("Mortalidad Infantil (por mil nacidos vivos)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
